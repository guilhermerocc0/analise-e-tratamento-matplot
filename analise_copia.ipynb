```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("dados_matriculas.xlsx")
```

```python
print(df)
```

```python
#verificar o numero de matriculas mensais
plt.figure(figsize=(10,5))

df["Matrículas"].plot(marker= "s", linestyle= "--", color="blue")
plt.title("Matrículas Mensais")
plt.xlabel("Mês")
plt.ylabel("Número de Matrículas")
plt.grid(True)
```

```python
# Identificar os meses com menor e maior numero de matriculas

max_matriculas = df["Matrículas"].max()
mes_max_matriculas = df["Matrículas"].idxmax()
print(f"Mes com maior numero de Matriculas {mes_max_matriculas}: {max_matriculas }matrículas")

#----------------------------------

min_matriculas = df["Matrículas"].min()
mes_min_matriculas = df["Matrículas"].idxmin()
print(f"Mes com menor cumero de matriculas {mes_min_matriculas}: {min_matriculas} matrículas")

```

```python
media_matricula = df["Matrículas"].mean()
```

```python
# Plotar o grafico geral 
plt.figure(figsize=(10,5))
df["Matrículas"].plot(marker= "o", linestyle= "-", color='red', label='Matrículas', mfc='Black', markersize='6')

# Adcionar linhas de informação para maximo e minimo de matricula
plt.axhline(y=max_matriculas, color='blue', linestyle='--', linewidth=2, label=f"Maximo: {max_matriculas}")

plt.axhline(y=min_matriculas, color='black', linestyle='--', linewidth=2, label=f"Minimo: {min_matriculas}")

plt.axhline(y=media_matricula, color='orange', linestyle='--', linewidth=2, label=f"Média: {media_matricula}")

plt.legend()
plt.title("Matrículas Mensais")
plt.xlabel("Mês")
plt.grid(True)
```

Tendência de Matrículas

Calcular a media matriculas por trimestr

QE = Trismetre
QS = Semestre
Q = Quadrimestre
A = Anual
D = Diario
W = Semanal
M = Mensal
H = horario

```python
# Passo - 1: Converter para Datetime 
df["Mês"] = pd.to_datetime(df["Mês"])

# Passo - 3: Definir a coluna de datas como indices
df.set_index("Mês", inplace=True)
# Passo - 3: Calcular media trimestral
matriculas_trimestre = df["Matrículas"].resample("QE").mean()

print(matriculas_trimestre)
```

```python
plt.figure(figsize=(10,5))
matriculas_trimestre.plot(kind='bar', color='blue')
plt.title("Média de Matrículas por trimestre")
plt.xlabel("Trimestre")
plt.ylabel("Media de Matrículas")
plt.xticks(rotation=45)
plt.grid(axis='y')
```
